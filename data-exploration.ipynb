```python
import kagglehub
from seaborn import heatmap

# Download latest version
path = kagglehub.dataset_download("dylanjcastillo/7k-books-with-metadata")

print("Path to dataset files:", path)
```

```python
import pandas as pd
```

```python
books = pd.read_csv(f"{path}/books.csv")
```

```python
books
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
ax = plt.axes()
sns.heatmap(books.isna().transpose(), cbar=False, ax=ax)

plt.xlabel("Columns")
plt.ylabel("Missing Values")

plt.show()
```

```python
import numpy as np

# Books where the description is missing ? put 1 in that column : put 0 in that column
books["missing_description"] = np.where(books["description"].isna(), 1, 0)

# Convert the year of the book to the age of the book
books["age_of_book"] = 2025 - books["published_year"]
```

```python
columns_of_interest = ["num_pages", "age_of_book", "missing_description", "average_rating"]

# Pearson is more suitable for continuous values, and Spearman is for non-continuous values
correlation_matrix = books[columns_of_interest].corr(method = "spearman")

# Convert the correlation matrix to the seaborn heat map
sns.set_theme(style="white")
plt.figure(figsize=(8, 6))
heatmap = sns.heatmap(correlation_matrix, annot=True, fmt=".2f", cmap="coolwarm",
                      cbar_kws={"label": "Spearman Correlation"})

heatmap.set_title("Correlation Heatmap")
plt.show()
```

```python
# Check if we can delete the missing data
books[(books["description"].isna()) |
      (books["average_rating"].isna()) |
      (books["num_pages"].isna()) |
      (books["published_year"].isna())]
```

```python
# Given the size of the dataset, less than 5% of data has missing fields
# Delete those missing data
book_missing = books[(~books["description"].isna()) &
      ~(books["average_rating"].isna()) &
      ~(books["num_pages"].isna()) &
      ~(books["published_year"].isna())]

book_missing
```

```python
 # Handle Category: There are too many category
# Let's see the statistics of the data category
book_missing["categories"].value_counts().reset_index().sort_values("count", ascending=False)
```

```python
# Category needs to be meaningful

# Count the number of words in the description
book_missing.loc[:, "words_in_description"] = book_missing["description"].str.split().str.len()
```

```python
book_missing
```

```python
# Find the cutoff point of words_in_description field to remove meaningless data

# Find the cutoff point
book_missing.loc[book_missing["words_in_description"].between(1, 4), "description"]
```

```python
# Find the cutoff point
book_missing.loc[book_missing["words_in_description"].between(25, 34), "description"]
```

```python
# Remove all of those which have less than 25 words in the description.
book_missing_25_words = book_missing[book_missing["words_in_description"] >= 25]
```

```python
book_missing_25_words
```

```python
# Create a new field called: title_and_subtitle

# When subtitle is missing, just use the title.
# If subtitle exists, use both title and subtitle
book_missing_25_words.loc[:, "title_and_subtitle"] = np.where(
    book_missing_25_words["subtitle"].isna(),
    book_missing_25_words["title"],
    book_missing_25_words[["title", "subtitle"]].astype(str).agg(": ".join, axis=1)
)
```

```python
book_missing_25_words
```

```python
# Create a new description column which tags each description with a unique identifier
# It will be important to build the vector database

book_missing_25_words.loc[:, "tagged_description"] = book_missing_25_words[["isbn13", "description"]].astype(str).agg(": ".join, axis=1)
```

```python
book_missing_25_words
```

```python
# Drop columns that we don't need
(
    book_missing_25_words.drop(["subtitle", "missing_description", "age_of_book", "words_in_description"], axis=1)
                        .to_csv("book_cleaned.csv", index=False)
)
```
